# **Career Prediction – Exploratory Data Analysis (EDA)**
This notebook includes 8 meaningful and essential EDA visualizations.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud

sns.set(style='whitegrid')

df = pd.read_csv('./../data/final_training_dataset.csv')
df.head()
```

## **1. Target Job Role Distribution**

```python
plt.figure(figsize=(10,5))
sns.countplot(data=df, x="Target Job Role", palette="viridis")
plt.title("Target Job Role Distribution")
plt.xticks(rotation=45)
plt.show()
```

## **2. Technical Skill Proficiency per Role**

```python
plt.figure(figsize=(10,5))
sns.barplot(data=df, x="Target Job Role", y="Tech Skill Proficiency", palette="coolwarm")
plt.title("Technical Skill Proficiency per Role")
plt.xticks(rotation=45)
plt.show()
```

## **3. Experience Months by Role**

```python
plt.figure(figsize=(10,5))
sns.boxplot(data=df, x="Target Job Role", y="Experience Months", palette="pastel")
plt.title("Experience Distribution per Role")
plt.xticks(rotation=45)
plt.show()
```

## **4. Correlation Heatmap of Numeric Features**

```python
numeric_df = df.select_dtypes(include=[np.number])
plt.figure(figsize=(12,10))
sns.heatmap(numeric_df.corr(), cmap='coolwarm', annot=False)
plt.title("Correlation Heatmap – Numeric Features")
plt.show()
```

## **5. Learning Hours vs Project Count**

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x="Total Hours Learning", y="Project Count", hue="Target Job Role", palette="tab10")
plt.title("Learning Hours vs Project Count")
plt.show()
```

## **6. Academic Scores Distribution (Class 10 & 12)**

```python
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
sns.histplot(df["Class 10 Percentage"], kde=True, bins=20, color="blue")
plt.title("Class 10 Percentage")

plt.subplot(1,2,2)
sns.histplot(df["Class 12 Percentage"], kde=True, bins=20, color="green")
plt.title("Class 12 Percentage")

plt.tight_layout()
plt.show()
```

## **7. Soft Skill Proficiency per Role**

```python
plt.figure(figsize=(10,5))
sns.barplot(data=df, x="Target Job Role", y="Soft Skill Proficiency", palette="magma")
plt.title("Soft Skill Proficiency per Role")
plt.xticks(rotation=45)
plt.show()
```

## **8. Word Cloud – Technical Skills**

```python
all_tech = " ".join(df["Technical Skills"].astype(str))
wordcloud = WordCloud(width=1200, height=600, background_color="white").generate(all_tech)

plt.figure(figsize=(12,6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.title("Technical Skills Word Cloud")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

df_comp = pd.read_csv("./../reports/model_comparison_results.csv")
df_comp
```

```python
plt.figure(figsize=(10,5))
sns.barplot(data=df_comp, x="Model", y="Accuracy", palette="viridis")
plt.title("Model Comparison – Accuracy")
plt.xticks(rotation=45)
plt.ylim(0,1)
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.barplot(data=df_comp, x="Model", y="Macro F1", palette="plasma")
plt.title("Model Comparison – Macro F1 Score")
plt.xticks(rotation=45)
plt.ylim(0,1)
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.barplot(data=df_comp, x="Model", y="Weighted F1", palette="coolwarm")
plt.title("Model Comparison – Weighted F1 Score")
plt.xticks(rotation=45)
plt.ylim(0,1)
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.plot(df_comp["Model"], df_comp["Accuracy"], marker="o", label="Accuracy")
plt.plot(df_comp["Model"], df_comp["Macro F1"], marker="o", label="Macro F1")
plt.plot(df_comp["Model"], df_comp["Weighted F1"], marker="o", label="Weighted F1")

plt.legend()
plt.title("Model Performance Trends")
plt.xticks(rotation=45)
plt.ylim(0,1)
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.heatmap(df_comp.set_index("Model"), annot=True, cmap="YlGnBu")
plt.title("Model Comparison Heatmap")
plt.show()
```
