```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
df = pd.read_csv("./../data/final_training_dataset.csv")
df.head()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(data=df, x="Target Job Role", palette="viridis")
plt.title("Distribution of Target Job Roles")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df["Tech Skill Proficiency"], kde=True, bins=20, color="blue")
plt.title("Technical Skill Proficiency Distribution")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df["Soft Skill Proficiency"], kde=True, bins=20, color="green")
plt.title("Soft Skill Proficiency Distribution")
plt.show()
```

```python
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
sns.histplot(df["Class 10 Percentage"], bins=20, kde=True)
plt.title("Class 10 Percentage")

plt.subplot(1,2,2)
sns.histplot(df["Class 12 Percentage"], bins=20, kde=True)
plt.title("Class 12 Percentage")

plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.barplot(data=df, x="Target Job Role", y="Tech Skill Proficiency", palette="coolwarm")
plt.title("Average Technical Skill Proficiency per Role")
plt.xticks(rotation=45)
plt.show()
```
